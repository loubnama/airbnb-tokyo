```python
import pandas as pd
import numbers as np
demand1=df['number_of_reviews_ltm'].mean()
print(demand1)
over1=df['price'].mean()
print(over1)
ava=df['availability_365'].mean()
print(ava)
avamax=df['availability_365'].max()
print(avamax)
```

```python
#Over يدل على السعر الغالي
#Demand يدل على الطلب
df['demand']=df['number_of_reviews_ltm']>=15
df['over']=df['price']>30000
df['ava_100']=df['availability_365']<=100
df['ava_200']=(df['availability_365']>100)&(df['availability_365']<=200)
df['ava_300']=(df['availability_365']>200)&(df['availability_365']<=300)
df['ava_400']=(df['availability_365']>300)&(df['availability_365']<=400)
df
```

```python
#أبحث عن القيم المفقودة
print(df.isnull().sum())
```

```python
df['last_review']=df['last_review'].fillna(0)
df['reviews_per_month']=df['reviews_per_month'].fillna(0)
df['price']=df['price'].fillna(0)
print(df.isnull().sum())
```

```python
#عمود اسمه الانتاجية يعطيكي قيمة صح اذا كان الريفيو للسنة الاخيرة للمضيف أكبر من متوسط الريفيو لكل سنة
df['yield']=df['number_of_reviews_ltm']>(df['number_of_reviews']/(df['number_of_reviews']/(df['reviews_per_month']*12)))

df
```

```python
#اصبح بامكاني حذف الاعمدة التي أصبحت غير ضرورية بعد انشاء عمود yield
df=df.drop(['availability_365','number_of_reviews_ltm','number_of_reviews','reviews_per_month'],axis=1)
df
```

```python
#اذا نجد ان عدد المضيف الذي إنتاجية السنة الأخيرة لديه اكبر من المعدل المتوسط هو 4813
df['yield'].value_counts()
```

```python
#مقارنة بين الإنتاجية والسعر 
#نلاحظ ان عندما يصبح الاسعار مرتفعة تصبح الإنتاجية الصحيحة معدومة
sns.catplot(x='price',col='yield',data=df,kind='box')
```

```python
#نلاحظ مع زيادة عرض الغرفة نجد إنتاجية صحيحة 
#ونلاحظ في انخفاض عرض الغرفة تكون الإنتاجية خاطئة يوجد منها القليل
sns.catplot(x='latitude',col='yield',data=df,kind='box')
```

```python
#ونلاحظ عند انخفاض طول الغرفة تكون ايضا الإنتاجية بقيمة خطأ
sns.catplot(x='longitude',col='yield',data=df,kind='box')
```

```python
#نلاحظ ان غرف المشاركة والفندق انتاجيتها منخفضة عن غيرها
sns.countplot(x='room_type',hue='yield',data=df)
```
